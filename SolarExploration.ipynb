```python
x = RegionsOnly.columns
colors=cm.rainbow(np.linspace(0,1,7))
for index in range(len(RegionsOnly.index)):
    y=RegionsOnly.iloc[index]
    plt.plot (x, y, c = colors[index], label = RegionalSolarGenerationDF.iloc[index], linewidth=4)
    plt.title("Regional Solar Generation")
    plt.xlabel("Years")
    plt.xticks(rotation=90)
    plt.ylabel("TeraWatt-hours")
    plt.legend(labels=CleanRegionalSolarDF.index)
    plt.savefig("Regional Solar Generation.png")
```
