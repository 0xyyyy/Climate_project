```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
from RenewablesExploration import WindGenerationDF
```

```python
WindGenerationDF
```

```python
RegionalWindGenerationDF = WindGenerationDF[WindGenerationDF.index.str.startswith("Total")]
CleanRegionalWindDF=RegionalWindGenerationDF.drop(WindGenerationDF.columns[0:45], axis=1)
RegionsOnly=CleanRegionalWindDF.drop(index="Total World")
RegionsOnly
```

```python
x = RegionsOnly.columns
colors=cm.rainbow(np.linspace(0,1,7))
for index in range(len(RegionsOnly.index)):
    y=RegionsOnly.iloc[index]
    plt.plot (x, y, c = colors[index], label = RegionsOnly.iloc[index], linewidth=4)
    plt.title("Regional Solar Generation")
    plt.xlabel("Years")
    plt.xticks(rotation=90)
    plt.ylabel("TeraWatt-hours")
    plt.legend(labels=RegionsOnly.index)
    plt.savefig("Regional Wind Generation.png")
```

```python
RegionDF = RegionsOnly.transpose()
PctChgDF=RegionDF.pct_change()
GraphVals=PctChgDF*100
GraphVals
```

```python
# group bar chart:
index=np.arange(10)
pos=list(range(len(GraphVals['Total North America'])))
barwidth=0.1428571429
groupgap=0.1428571429
labels=["Total North America", "Total S. & Cent. America", "Total Europe", "Total CIS", "Total Middle East", "Total Africa", "Total Asia Pacific"]

fig, ax = plt.subplots(figsize=(20, 10))
rects1=ax.bar(pos, GraphVals["Total North America"], barwidth, color=colors[0])
rects2=ax.bar([p + barwidth for p in pos], GraphVals["Total S. & Cent. America"], barwidth, color=colors[1])
rects3=ax.bar([p + barwidth*2 for p in pos], GraphVals["Total Europe"], barwidth, color=colors[2])
rects4=ax.bar([p + barwidth*3 for p in pos], GraphVals["Total CIS"], barwidth, color=colors[3])
rects5=ax.bar([p - barwidth for p in pos], GraphVals["Total Middle East"], barwidth, color=colors[4])
rects6=ax.bar([p - barwidth*2 for p in pos], GraphVals["Total Africa"], barwidth, color=colors[5])
rects7=ax.bar([p - barwidth*3 for p in pos], GraphVals["Total Asia Pacific"], barwidth, color=colors[6])
plt.xticks(index, GraphVals.index)
plt.title("Annual Growth Rate of Renewable Energy Generation")
plt.ylabel("% Change")
plt.legend(labels=labels)
plt.savefig("AGRWindGeneration.png")
plt.show()
```
