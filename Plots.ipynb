```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from RenewablesExploration import RenCon_df, RenGen_df
```

```python
TotalRegion_df=RenCon_df[RenCon_df.index.str.startswith("Total")]
TotalRegion_df.index.name = "Region"
RegionalPlotData=TotalRegion_df.drop(["2019.2", "2008-18", "2019.1"], axis=1)
```

```python
Transposed=RegionalPlotData.transpose().drop(index=["Unnamed: 59", "Unnamed: 60"])
```

```python
# Convert values to arrays for matplotlib intake
x=np.asarray(Transposed.index)
y=np.asarray(Transposed.values)
# For each Region(column) plot a separate line based on the numpy arrays, format plot
for column in RegionalPlotData:
    plt.plot(x, y, marker='', linewidth=1, alpha=0.9, label=column)
    plt.title("Renewable Energy Consumption Through The Years")
    plt.xlabel("Years")
    plt.ylabel("Consumption (Exajoules)")
    plt.legend(Transposed.columns)
    plt.savefig("YearlyRegionalConsumption.png")
```

```python
Main=RenCon_df[:-3].copy().drop(RegionalPlotData.index).drop(["2019.2", "2008-18", "2019.1"], axis=1).drop(["Unnamed: 59", "Unnamed: 60"], axis=1) 
Main
```

```python
AncientHistory=Main[Main.columns[:-20]]
RecentHistory=Main[Main.columns[-20:]]
RecentHistory.describe()
```

```python
AncientHistory.describe()
```

```python
# Find the top 10 consumers of renewable energies in 2019
Transposed2 = Main.transpose()
TopUsers = Transposed2.iloc[Transposed2.index == 2019].transpose().sort_values(by=[2019], ascending=False).head(10)
TopUsers
```

```python
CNames=np.asarray(TopUsers.index)
Values=np.asarray(TopUsers.values)
def Extract(Values):
    return [item[0] for item in Values]
    print(Extract(Values))
Values=Extract(Values)
plt.bar(CNames,Values)
plt.title("Top Consumers of Renewable Energy in 2019")
plt.ylabel("Exojoules")
plt.xticks(rotation='90')
plt.show()
plt.savefig("TopConsumers2019.png")
```

```python
# Find the top 10 consumers of renewable energies in 1999
Transposed2 = Main.transpose()
TopUsers2 = Transposed2.iloc[Transposed2.index == 1999].transpose().sort_values(by=[1999], ascending=False).head(10)
TopUsers2
```

```python
MeanNMax=pd.DataFrame({"Mean":Main.mean(), "Max":Main.max()})
plt.plot(MeanNMax)
plt.title("Annual Max vs Mean Global Growth")
plt.show()
plt.savefig("Annual Global Consumption Growth Max vs Mean")
```
